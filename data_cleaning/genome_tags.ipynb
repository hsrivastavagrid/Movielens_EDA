```python
import pandas as pd
import numpy as np

genome_tags = pd.read_csv('../data/raw_data/genome-tags.csv')
```

```python
genome_tags.head()
genome_tags.shape
genome_tags.info()
```

```python
genome_tags.isna().sum()
```

```python
genome_tags = genome_tags.dropna(subset=["tag"])
```

```python
genome_tags["tagId"].duplicated().sum()
```

```python
genome_tags["tag_clean"] = (
    genome_tags["tag"]
    .str.lower()
    .str.strip()
)
```

```python
genome_tags["tag_clean"] = genome_tags["tag_clean"].str.replace(
    r"[^\w\s]", "", regex=True
)
```

```python
genome_tags["tag_clean"].duplicated().sum()
```

```python
genome_tags["tag_clean"].str.len().describe()
```

```python
genome_tags.to_csv(
    "../data/cleaned_data/genome_tags_clean.csv",
    index=False
)
```

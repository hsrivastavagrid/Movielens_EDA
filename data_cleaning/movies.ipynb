```python
import pandas as pd
import numpy as np

movies = pd.read_csv('../data/raw_data/movies.csv')
```

```python
movies.head()
movies.shape
movies.info()
```

```python
movies.isna().sum()
```

```python
movies["movieId"].duplicated().sum()
```

```python
#Extracting release year from title
movies["release_year"] = (
    movies["title"]
    .str.extract(r"\((\d{4})\)")
    .astype("float")
)
```

```python
movies["release_year"].describe()
```

```python
movies["clean_title"] = (
    movies["title"]
    .str.replace(r"\s*\(\d{4}\)", "", regex=True)
)
```

```python
movies["genre_list"] = movies["genres"].str.split("|")
```

```python
movies["genre_list"].apply(
    lambda x: "(no genres listed)" in x
).sum()
```

```python
movies_genres = movies.explode("genre_list")

movies_genres = movies_genres[
    movies_genres["genre_list"] != "(no genres listed)"
]
```

```python
movies_genres["genre_list"].value_counts().head(10)
```

```python
movies.to_csv("../data/cleaned_data/movies_clean.csv", index=False)
movies_genres.to_csv("../data/cleaned_data/movies_genres_clean.csv", index=False)
```

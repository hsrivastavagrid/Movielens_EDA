```python
import pandas as pd
import numpy as np

ratings = pd.read_csv('../data/raw_data/ratings.csv')
```

```python
ratings.head()
ratings.shape
ratings.info()
```

```python
ratings.isna().sum()
```

```python
ratings["rating"].describe()
```

```python
assert ratings["rating"].between(0.5, 5.0).all(), "Invalid ratings detected"
```

```python
ratings.duplicated(
    subset=["userId", "movieId", "timestamp"]
).sum()
```

```python
ratings["timestamp"] = pd.to_datetime(
    ratings["timestamp"], unit="s"
)
```

```python
ratings["timestamp"].min(), ratings["timestamp"].max()
```

```python
ratings["timestamp"].dt.year.describe()
```

```python
ratings["rating_year"] = ratings["timestamp"].dt.year
ratings["rating_month"] = ratings["timestamp"].dt.to_period("M")
```

```python
ratings["userId"].nunique()
ratings.groupby("userId").size().describe()
```

```python
ratings["movieId"].nunique()
ratings.groupby("movieId").size().describe()
```

```python
ratings.to_csv(
    "../data/cleaned_data/ratings_clean.csv",
    index=False
)
```
